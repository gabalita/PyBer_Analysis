```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create a Line Chart Using the MATLAB Method
plt.plot(x_axis,y_axis)
```

```python
# Create a Line Chart Using the Object-Oriented Interface, ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis,y_axis)

# The plt.subplots() function returns a tuple that contains a figure and axes object(s).
```

```python
# Method 2
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis,y_axis)
```

```python
# Method 3
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()

# plt.show() should only be used once per session because it can cause the graph to display improperly. If you
# have problems with data not being displayed properly, you may have to restart your kernel and clear your output.
```

```python
# Create the plot
plt.plot(x_axis,y_axis,label="Boston",marker='*',color = 'blue', linewidth = 2)

# Create the labels
plt.xlabel('Time')
plt.ylabel('Total Fare Accumulated')

# Create axis limits and title and legend
plt.ylim(0,45)
plt.title('PyBer Fare Over Month')
plt.legend()

plt.figure(figsize=(60,40))

# To limit x-axis, you could do:
# plt.xlim('Jan','June')
# plt.xlim(0,5)
```

```python
# Create the plot 
fig, ax = plt.subplots(figsize=(10,5))

# Assign the data to the plot
ax.plot(x_axis,y_axis,label='Boston',color='green', linewidth = 2, marker='*')

# Set y Limit
ax.set_ylim(0,45)

# Create title
ax.set_title('PyBer Fare Over Time')

# Add the legend
ax.legend()

# Add a grid
ax.grid()
```

```python
# Creating Bar Charts with MatLab

plt.bar(x_axis,y_axis, color='green', label='Boston')
plt.ylabel('Fare Total ($)')
plt.xlabel('Date')
plt.title('PyBer Fare Over Month')
plt.legend()
```

```python
# Creating Horizontal Bar Charts with MatLab

plt.barh(x_axis,y_axis, color='magenta', label='Boston')
plt.ylabel('Date')
plt.xlabel('Fare ($)')
plt.title('PyBer Fare Over Month')
plt.legend()
plt.gca().invert_yaxis()
```

```python
# Creating Bar Charts with Object Oriented Method ax.plt()

# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]


# Create the figure and axis objects
fig, ax = plt.subplots()

# Declare the ax object as a bar chart
ax.bar(x_axis, y_axis, color='black', label = 'New York')
ax.legend()
```

```python
# Create horizontal bar chart with OOI

fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color ='blue', label = "Chicago")
ax.legend()
ax.invert_yaxis()
```

```python
# Creating Scatter Plots with MatLab

# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

plt.plot(x_axis,y_axis,'o')
```

```python
plt.scatter(x_axis,y_axis)
```

```python
plt.scatter(y_axis,x_axis, label= "Chicago", color = 'red')
plt.title('Fare Over Time')
plt.xlabel('Fare ($)')
plt.ylabel('Date')
plt.xlim(0,50)
plt.gca().invert_yaxis()
```

```python
plt.scatter(x_axis,y_axis, s=y_axis)

y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
    
plt.scatter(x_axis,y_axis, s=y_axis_larger)
```

```python
# Refactoring code using list comprehension technique

plt.scatter(x_axis, y_axis, s = [3*i for i in y_axis])
```

```python
# Create a Scatter Plot Using the Object-Oriented Interface

fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s=[i*5 for i in y_axis], color = 'royalblue', label = 'Boston', linewidth=2, alpha=0.5, edgecolors='black')
ax.set(xlabel = 'Fare($)', ylabel = 'Date')
ax.set_title('Fare Over Time')
ax.set_xlim(0,50)
ax.legend()
ax.invert_yaxis()
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%', colors = colors)
plt.show()
```

```python
# Creating pie chart with OOI
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
fig, ax = plt.subplots()
ax.pie(y_axis, labels=x_axis, autopct = '%.f%%', colors = colors, 
       explode=explode_values, shadow = True, startangle = 90,
      counterclock = False)
plt.show()
```
